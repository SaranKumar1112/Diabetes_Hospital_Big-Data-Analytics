```python
import pandas as pd

# Load dataset
df = pd.read_csv('diabetic_data.csv')

# specific columns
columns_chosen = [
    'time_in_hospital', 
    'num_lab_procedures', 
    'num_medications', 
    'number_diagnoses', 
    'discharge_disposition_id', 
    'number_outpatient',
    'age'
]

df_select = df[columns_chosen]


print(df_select.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.boxplot(x='time_in_hospital', y='age', data=df, palette='Set2')
plt.title('Distribution of Time in Hospital by Age', fontsize=16)
plt.xlabel('Time in Hospital (Days)', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#  2x2 grid box plots
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# num_medications
sns.boxplot(x='num_medications', y='age', data=df, palette='Set2', ax=axs[0, 0])
axs[0, 0].set_title('Age Distribution by Number of Medications', fontsize=14)
axs[0, 0].set_xlabel('Number of Medications', fontsize=12)
axs[0, 0].set_ylabel('Age', fontsize=12)

# num_lab_procedures
sns.boxplot(x='num_lab_procedures', y='age', data=df, palette='Set2', ax=axs[0, 1])
axs[0, 1].set_title('Age Distribution by Number of Lab Procedures', fontsize=14)
axs[0, 1].set_xlabel('Number of Lab Procedures', fontsize=12)
axs[0, 1].set_ylabel('Age', fontsize=12)

# number_diagnoses
sns.boxplot(x='number_diagnoses', y='age', data=df, palette='Set2', ax=axs[1, 0])
axs[1, 0].set_title('Age Distribution by Number of Diagnoses', fontsize=14)
axs[1, 0].set_xlabel('Number of Diagnoses', fontsize=12)
axs[1, 0].set_ylabel('Age', fontsize=12)

# time_in_hospital
sns.boxplot(x='time_in_hospital', y='age', data=df, palette='Set2', ax=axs[1, 1])
axs[1, 1].set_title('Age Distribution by Time in Hospital', fontsize=14)
axs[1, 1].set_xlabel('Time in Hospital (Days)', fontsize=12)
axs[1, 1].set_ylabel('Age', fontsize=12)


plt.tight_layout()
plt.suptitle('Age Distribution by Various Factors', fontsize=18, y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2x2 grid for histograms 
fig, axs = plt.subplots(2, 2, figsize=(14, 10))


hist_params = {
    'bins': 30,
    'edgecolor': 'black',
    'alpha': 0.7
}

# time_in_hospital
sns.histplot(df['time_in_hospital'], bins=30, color='skyblue', kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Time in Hospital', fontsize=14)
axs[0, 0].set_xlabel('Days in Hospital', fontsize=12)
axs[0, 0].set_ylabel('Frequency', fontsize=12)

#for num_lab_procedures
sns.histplot(df['num_lab_procedures'], bins=30, color='salmon', kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Distribution of Number of Lab Procedures', fontsize=14)
axs[0, 1].set_xlabel('Number of Lab Procedures', fontsize=12)
axs[0, 1].set_ylabel('Frequency', fontsize=12)

# num_medications
sns.histplot(df['num_medications'], bins=30, color='lightgreen', kde=True, ax=axs[1, 0])
axs[1, 0].set_title('Distribution of Number of Medications', fontsize=14)
axs[1, 0].set_xlabel('Number of Medications', fontsize=12)
axs[1, 0].set_ylabel('Frequency', fontsize=12)

# number_diagnoses
sns.histplot(df['number_diagnoses'], bins=30, color='lightcoral', kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Distribution of Number of Diagnoses', fontsize=14)
axs[1, 1].set_xlabel('Number of Diagnoses', fontsize=12)
axs[1, 1].set_ylabel('Frequency', fontsize=12)


plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# correlation matrix calculation
correlation_matrix = df[['time_in_hospital', 'num_lab_procedures', 'num_medications', 'number_diagnoses']].corr()

print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```

```python
import pandas as pd


variables = ['time_in_hospital', 'num_lab_procedures', 'num_medications', 'number_diagnoses', 'number_outpatient']


correlation_matrix = df[variables].corr()


print("Correlation Matrix:")
print(correlation_matrix)

# heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Heatmap of Key Variables', fontsize=16)
plt.show()

```

```python
import statsmodels.api as sm
import pandas as pd




y = df['time_in_hospital']


X = df[['num_lab_procedures', 'num_medications', 'number_diagnoses', 'number_outpatient']]


X = sm.add_constant(X)

model = sm.OLS(y, X).fit()


print(model.summary())
```
